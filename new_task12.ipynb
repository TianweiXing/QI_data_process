```python
import os
import numpy as np
import pandas as pd
```

# Task 1

```python
def extract_task1(df, monitor = True):
    
    # change gp_co to str
    # remove "."
    # change "nan" to " "
    
    # print(df.SCAG_GP_CO.unique().shape)
    # print(df.SCAG_GP_CO.unique())
    df.SCAG_GP_CO = df.SCAG_GP_CO.astype('str')
    df.SCAG_GP_CO[(df.SCAG_GP_CO == 'nan') ] = ' '
    df.SCAG_GP_CO = df.SCAG_GP_CO.str.split('.',expand=True)[0]
    # print(df.SCAG_GP_CO.unique().shape)
    # print(df.SCAG_GP_CO.unique())
    
    # calculating area and generating result mat
    unique_gp_co = df.SCAG_GP_CO.unique()
    if monitor == True:
        print(unique_gp_co)
        print('GP_CO size: ', unique_gp_co.shape)

    unique_city = df.CITY_NAME.unique()
    if monitor == True:
        print(unique_city)
        print('CITY size: ', unique_city.shape)

    area_sum_mat = np.zeros([unique_gp_co.shape[0], unique_city.shape[0] ])

    for id_ct, city_i in enumerate(unique_city):
        print('Current city: %d, \t %s.'%(id_ct, city_i))
        for id_co, gp_co_i in enumerate(unique_gp_co):
            data_i = df[(df.CITY_NAME == city_i) & (df.SCAG_GP_CO == gp_co_i) ]
            area_sum_mat[id_co, id_ct] = data_i.Area.sum()

    unique_gp_co = unique_gp_co.astype('str')
    unique_gp_co[np.where(unique_gp_co == 'nan')] = 'NoLa'  # not valid now after cleaning LA data
    unique_gp_co[np.where(unique_gp_co == ' ')] = 'NoLa'

    if monitor == True:
        print(unique_gp_co)

    extracted_df = pd.DataFrame(area_sum_mat, columns=unique_city, index = unique_gp_co)

    return extracted_df
```

```python
data_dir = 'data/'
org_files = os.listdir(data_dir)
print(org_files)


for file_i in org_files:
    path_i = data_dir+ file_i
    print(path_i)
    print('========================================')
    
    df = pd.read_csv(path_i)
    extracted_df = extract_task1(df)
    
    save_path_i = 'result_1/'+ file_i.split('.')[0] +'_TASK1.csv'
    print('Saving result to: ', save_path_i)
    extracted_df.to_csv(save_path_i)
    print('\n\n')
    
#     break
    
```

```python
# Task 2:
# use extracted matrix:
# for each city
# every big category
# calculate sum
# result:
# all cities in one matrix
# add one more feature, shows the county of the city.
```

```python
result_folder = 'result_1'
result1_file_list = os.listdir(result_folder)
print(result1_file_list)
```

```python
Single_Family_Residential = np.array([1110]).astype('str')
Multi_Family_Residential = np.array([1120]).astype('str')
Other_Residential = np.array([1100, 1130, 1140]).astype('str')
Commmercial = np.array([1210, 1200, 1220, 1230, 1240, 1250]).astype('str')
Industrial = np.array([1300, 1310, 1320, 1330, 1340, 1350]).astype('str')
Institutional_Public_Facilities = np.array([1260, 1270, 1280, 1290, 1400, 1410, 1420]).astype('str')
Open_Space = np.array([1800, 1810, 1820, 1830, 1840, 1850, 1860, 1870, 1880]).astype('str')
Mixed_Used = np.array([1500, 1600]).astype('str')
Agriculture = np.array([2000]).astype('str')
Other = np.array([4000]).astype('str')
Specific_Plan = np.array([7777]).astype('str')
Vacant = np.array([3000, 1900]).astype('str')
No_Label = np.array(["NoLa", 9999]).astype('str')
TBD = np.array([0]).astype('str')

big_category = [Single_Family_Residential, Multi_Family_Residential, Other_Residential,
               Commmercial, Industrial, Institutional_Public_Facilities, 
               Open_Space, Mixed_Used, Agriculture,
               Other, Specific_Plan, Vacant, No_Label, TBD]
print('In total %d big categories.'%len(big_category))
```

```python
def task2_process(file_path, county_name, big_category):

    file_name_path = file_path
    data = pd.read_csv(file_name_path)
    df = pd.DataFrame(data)
    df.rename(columns={'Unnamed: 0':'SCAG_GP_CO'}, inplace=True)
    
    city_list = df.columns[1:]
    print('\n There are in total %d cities in \"%s\": '%(city_list.shape[0], file_name_path))
    print(city_list)
    
    county_matrix = np.zeros([city_list.shape[0], len(big_category)])
    
    
    for idx_city, city_i in enumerate(city_list):
        for idx_cat, category_i in enumerate(big_category):
    #         print('Current_city: %s, Current category: %s'%(city_i, category_i))
            county_matrix[idx_city, idx_cat] = df[city_i][ np.isin(df.SCAG_GP_CO, category_i) ].sum()
    city_area = county_matrix.sum(axis = 1).reshape([-1,1])

    new_county_matrix = np.concatenate([county_matrix,city_area], axis =1)
#     print(new_county_matrix.shape)
    new_county_matrix = new_county_matrix.astype('str')
    
    city_name_column = np.zeros([city_list.shape[0], 1]).astype('str')
    city_name_column[:,0] = city_list
    result_matrix = np.concatenate([city_name_column, new_county_matrix], axis =1)
#     print(result_matrix.shape)
    
    county_name_column = np.zeros([city_list.shape[0], 1]).astype('str')
    county_name_column[:,:] = county_name
    result_matrix = np.concatenate([county_name_column, result_matrix], axis =1)
#     print(result_matrix.shape)
    
    return result_matrix
```

```python
result1_file_list
```

```python
result_list = []
for file_i in result1_file_list:
    file_name_path = result_folder +'/' + file_i
    county_name_i = file_i.split('_')[1]
    print('\n\n',county_name_i)
    result_list.append( task2_process(file_name_path, county_name_i, big_category) )
```

```python
final_result = np.concatenate([result_list[0], result_list[1],  result_list[2],  result_list[3],  result_list[4]], axis = 0)
final_result.shape
```
