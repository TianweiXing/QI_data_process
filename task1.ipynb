```python
file_name_path = 'org_files/LandUse_Combined_San_bernardino_.csv'
data = pd.read_csv(file_name_path)
df = pd.DataFrame(data)
unique_gp_co = df.SCAG_GP_CO.unique().astype('str')
unique_gp_co[np.where(unique_gp_co == 'nan')] = 'NoLa'
unique_gp_co[np.where(unique_gp_co == ' ')] = 'NoLa'
```

```python
def extract_task1(file_path, file_name, save_path):
    name = file_name.split('.')[0].split('_')

    file_name_path = file_path+'/'+file_name
    save_path = save_path + '/' + name[0]+'_'+name[2]+'_'+'deCat.csv'
    print('\n===================== ',file_name, '=================')
    data = pd.read_csv(file_name_path)
    df = pd.DataFrame(data)
    
    unique_gp_co = df.SCAG_GP_CO.unique()
#     print(unique_gp_co)
    print('GP_CO size: ', unique_gp_co.shape)

    unique_city = df.CITY.unique()
#     print(unique_city)
    print('CITY size: ', unique_city.shape)

    area_sum_mat = np.zeros([unique_gp_co.shape[0], unique_city.shape[0] ])

    for id_ct, city_i in enumerate(unique_city):
        print('Current city: %d, \t %s.'%(id_ct, city_i))
        for id_co, gp_co_i in enumerate(unique_gp_co):
            data_i = df[(df.CITY == city_i) & (df.SCAG_GP_CO == gp_co_i) ]
            area_sum_mat[id_co, id_ct] = data_i.ACRES.sum()
    
    unique_gp_co = unique_gp_co.astype('str')
    unique_gp_co[np.where(unique_gp_co == 'nan')] = 'NoLa'  # not valid now after cleaning LA data
    unique_gp_co[np.where(unique_gp_co == ' ')] = 'NoLa'

#     print(unique_gp_co)

    extracted_df = pd.DataFrame(area_sum_mat, columns=unique_city, index = unique_gp_co)

    extracted_df.to_csv(save_path)
```
